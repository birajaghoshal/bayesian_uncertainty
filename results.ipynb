```python
import pickle
from collections import namedtuple
import pandas as pd
import operator
from IPython.display import display
from metrics import Results
```

```python
metrics = 'normal_nll rmse auc_rmse'.split()

def highlight_max(s):
    is_min = s == s.min()
    return ['background-color: yellow' if v else '' for v in is_min]

with open('shallow_results.pkl', 'rb') as f:
    pr = pickle.load(f)
    
with open('deep_results.pkl', 'rb') as f:
    dr = pickle.load(f)
```

```python
df = pd.DataFrame.from_records(pr + dr, columns=Results._fields)
df = df[df.model != "GBTQuantile"]
```

```python
for m in metrics:
    print(m.upper())
    pdf = df.pivot(index='dataset', columns='model', values=m)
    pdf.columns = ["BayesianLR"] + sorted(pdf.columns)[1:]
    display(pdf.applymap(operator.itemgetter(0)).style.apply(highlight_max, axis=1))
```
