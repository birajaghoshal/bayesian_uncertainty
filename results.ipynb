```python
import pickle
from collections import namedtuple
import pandas as pd
import operator
from IPython.display import display
```

```python
Results = namedtuple('Results', 'datetime dataset model shape cv_metrics normal_nll rmse mae auc_rmse auc_mae')

metrics = 'normal_nll rmse mae auc_rmse auc_mae'.split()

def highlight_max(s):
    is_min = s == s.min()
    return ['background-color: yellow' if v else '' for v in is_min]

with open('par_results.pkl', 'rb') as f:
    pr = pickle.load(f)
```

```python
df = pd.DataFrame.from_records(pr, columns=Results._fields)
```

```python
for m in metrics:
    print(m.upper())
    pdf = df.pivot(index='dataset', columns='model', values=m)
    display(pdf.applymap(operator.itemgetter(0)).style.apply(highlight_max, axis=1))
```
