```python
%matplotlib inline
import time
import random
import datetime
from collections import OrderedDict
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import scipy
from tqdm import tqdm_notebook
from xgboost import XGBRegressor, XGBClassifier
import sklearn
from sklearn import linear_model
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import cross_val_score
from sklearn.metrics import roc_auc_score
from sklearn.base import BaseEstimator, RegressorMixin
```

```python
df = pd.read_csv('datasets/shuttle.txt', sep=' ', header=None)
```

```python
reg = XGBRegressor(n_estimators=300)
lr = linear_model.LinearRegression()
```

```python
df['label'] = df[9]-1
df['reward'] = np.nan
df['action'] = np.nan
df = df.drop(columns=[9], axis=1)
```

```python
history = pd.DataFrame()
log = []
acc_regret = 0
start_epsilon = 10
end_epsilon = 1000
first_time = True
actions = range(7)

for i in range(43500):
    sample = df.sample(n=1)
    action = None
    epsilon = lr.fit([[start_epsilon], [end_epsilon]], [1.0, 0.0]).predict(i)
    
    if random.random() < epsilon[0]:
        action = random.choice(actions)
    elif i % 100 == 0 or first_time:
        first_time = False
        X_train = history.iloc[-10000:].drop(['label', 'reward'], axis=1)
        y_train = history.iloc[-10000:]['reward']
        reg.fit(X_train, y_train)
    
    if action is None:
        pred = {}
        for a in actions:
            sample['action'] = a
            pred[a] = reg.predict(sample.drop(['label', 'reward'], axis=1))
        action = max(pred.keys(), key=(lambda k: pred[k]))

    label = sample['label'].values[0]

    reward = 0
    reward = int(action == label)
    uniform = int(random.choice(actions) == label)
    
    sample['action'] = action
    sample['reward'] = reward
    
    history = history.append(sample)
    
    log.append((i, action, reward, uniform))
    print(i, action, reward, uniform)
```

```python
results = pd.DataFrame(log, columns='i action reward uniform'.split())
```
