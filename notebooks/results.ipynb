```python
import pickle
from collections import namedtuple
import pandas as pd
import operator
import numpy as np
from IPython.display import display
from metrics import Results
import numbers
```

```python
metrics = 'train_time normal_nll normal_nll_opt rmse auc_rmse auc_rmse_norm'.split()

def highlight_max(s):
    is_min = s == s.min()
    return ['background-color: yellow' if v else '' for v in is_min]

with open('shallow_results.pkl', 'rb') as f:
    pr = pickle.load(f)
    
#with open('deep_results.pkl', 'rb') as f:
#    dr = pickle.load(f)
```

```python
df = pd.DataFrame.from_records(pr, columns=Results._fields)
df
```

```python
for m in metrics:
    print(m.upper())
    pdf = df.pivot(index='dataset', columns='model', values=m)
    pdf = pdf.rename({"BayesianLinearRegression":"BayesianLR", "LinearRegression":"LR", "MLPBayesianDropout": "BayesianMLP"}, axis=1)
    def getter(cell):
        if isinstance(cell, numbers.Number):
            return cell
        else:
            try:
                return cell[0]
            except:
                return np.nan
    display(pdf.applymap(getter).style.apply(highlight_max, axis=1))
```

```python
pdf = df.pivot(index='dataset', columns='model', values='auc_rmse')
pdf = pdf.rename({"BayesianLinearRegression":"BayesianLR", "LinearRegression":"LR", "MLPBayesianDropout": "BayesianMLP"}, axis=1)
def getter(cell):
    if isinstance(cell, numbers.Number):
        return cell
    else:
        try:
            return cell[0]
        except:
            return np.nan
a=pdf.applymap(getter)
```

```python
rows = ["\\begin{tabular}{llllllll}\n\\toprule"]
rows.append("Datasets & " + " & ".join(pdf.columns) + " \\\\")
rows.append("\midrule")

n_bests = []

for d, m in pdf.iterrows():
    means = np.array([i[0] for i in m])
    stds = np.array([i[1] for i in m])
    best = np.min(means)
    best_std = stds[np.argmin(means)]
    if np.isfinite(best_std): 
        bests = (means - stds) <= best + best_std
    else:
        bests = means == best
    n_bests.append(bests)
    row = ["\\textbf{{ {0:.2f} \\tpm {1:.2f} }}".format(m, s)  if b else "{0:.2f} \\tpm {1:.2f}".format(m, s) for m,s,b in zip(means, stds, bests)]
    rows.append(d.replace('_', '\\textunderscore ') + " & " + " & ".join(row) + " \\\\")
    
n_bests = np.sum(n_bests, axis=0)
rows.append("\\midrule")
rows.append("Number of Bests & " + " & ".join([str(i) for i in n_bests]) + " \\\\")
rows.append("\\bottomrule\n\\end{tabular}\n")
print('\n'.join(rows))
```
