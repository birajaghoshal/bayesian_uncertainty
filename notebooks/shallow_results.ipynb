```python
def highlight_max(s):
    is_min = s == s.min()
    return ['background-color: yellow' if v else '' for v in is_min]
```

```python
cols = [c for c in df.model.unique() if not c.startswith("UncertaintyCalibrator")]
uc_cols = {c : c.split('_')[1] for c in df.model.unique() if c.startswith("UncertaintyCalibrator")}
```

```python
df.pivot(index="dataset", columns="model", values="nlpd")[cols].style.apply(highlight_max, axis=1)
```

```python
df.pivot(index="dataset", columns="model", values="nlpd")[list(uc_cols.keys())].rename(uc_cols, axis=1).style.apply(highlight_max, axis=1)
```
